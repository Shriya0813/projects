```python
#Theatre ticket booking system-

price_ruby = 200
price_emerald = 250
price_diamond = 400
price_premium = 500
popcorn = 150
cold_drink = 150
small_combo = 250
large_combo = 400
total_seats = 150
occupied_seats = 0

while occupied_seats < total_seats:
    total = 0
    name = input("Enter your name: ")
    
    print("""Press 1 for ruby seats (price- 200/-) 
    Press 2 for emerald seats (price- 250/-)
    Press 3 for diamond seats (price- 400/-)
    Press 4 for premium seats (price- 500/-)""")

    choice = int(input("Enter your choice between 1/2/3/4: "))

    if choice in (1, 2, 3, 4):
        seats = int(input("Enter number of seats: "))
        if seats + occupied_seats <= total_seats:
            snacks = input("Do you want to add snacks?yes/no: ")
            if choice == 1:
                total = seats*price_ruby
            elif choice == 2:
                total = seats*price_emerald
            elif choice == 3:
                total = seats*price_diamond
            elif choice == 4:
                total = seats*price_premium

            if snacks == "yes":
                print("""Press 1 for popcorn (price- 150/-)
                Press 2 for cold drink (price- 150/-)
                Press 3 for small combo (price- 250/-)
                Press 4 for large combo (price- 400/-)""")
                items = int(input("Enter your choice between 1/2/3/4: "))
                if items == 1:
                    total += popcorn
                elif items == 2:
                    total += cold_drink
                elif items == 3:
                    total += small_combo
                elif items == 4:
                    total += large_combo
                    
            occupied_seats += seats
        else:
            print("Enough seats not available")
            break

    else:
        print("Invalid input...try again")
        break

    print("-"*30)
    print("    ***Bill***")
    print("Customer's name: ",name)
    print("Total amount: ", total)
    print("-"*30)

    next_person = input("Next person? yes/no: ")
    available_seats = total_seats - occupied_seats
    print("Number of seats left: ",available_seats)
    if next_person == "no" or occupied_seats >= total_seats:
        break

if occupied_seats >= total_seats:
    print("House Full")
```
